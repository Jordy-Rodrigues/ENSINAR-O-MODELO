```python
# ABRIR A PLANILHA COM O PANDAS NO GOOGLE COLAB
data=pd.read_excel('/content/drive/MyDrive/dataset_tablet_smartphone.xlsx')
```

```python
# ABRIR A PLANILHA 
data
```

```python
# VER DETALHES DA TABELA
data.dtypes
```

```python
# MODIFICAR A COLUNA PORTATIL DO TIPO OBJETO PARA NUMÉRICA TIPO INTEIRO
converter = {"Portatil": {"Smartphone":1,"Tablet":2}}
data.replace(converter,inplace=True) # inplace=True, FAZ COM QUE A ALTERAÇÃO SEJA FEITA NA PRÓPRIA TABELA ORIGINAL

# MOSTRAR O CABEÇALHO E AS 5 PRIMEIRAS LINHAS 
data.head()
```

```python
# SEPARAR A VARIÁVEL X DEPENDENTE E A VARIÁVEL Y INDEPENDENTE
x = data.drop (columns=['Portatil']) # TUDO EXCETO A COLUNA 'Portatil'
y = data.Portatil # RECEBE A COLUNA 'Portatil'
```

```python
print(y)
```

```python
x
```

```python
# IMPORTAR AS BIBLIOTECAS DO SKLEARN
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# SEPARAR UMA PARTE DA BASE DE DADOS PARA TREINAR E TESTAR O MODELO
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.4)
```

```python
# CHAMAR O CLASSIFICADOR DE ÁRVORE DE DECISÃO E TREINAR A INTELIGÊNCIA ARTIFICIAL
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
# TESTAR O MODELO COM NOVOS DADOS QUE ELE NÃO VIU
y_pred = clf.predict(x_test) # 40% NÃO UTILIZADOS DA VARIÁVEL 'clf'
y_true = y_test # GABARITO COM AS RESPOSTAS

# COMPARAÇÃO DE ACERTOS
print("Resultados obtidos: ")
print(y_pred)
print("\n Resultados esperados: ")
print(y_true.values)
```

```python
# IMPORTAR A BIBLIOTECA MÉTRICAS DO SKLEARN
from sklearn import metrics
# MOSTRAR A PRECISÃO DO MODELO PASSANDO O GABARITO E AS RESPOSTAS
print("precision"+str(metrics.precision_score(y_true,y_pred)))
```
